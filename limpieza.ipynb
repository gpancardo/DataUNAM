## Reducir t√≠tulos
Recortar elementos redundantes de los t√≠tulos (__'Career'__) para simplificar lectura. Esta informaci√≥n tambi√©n est√° disponible en el atributo __'Academic Degree'__.

```python
df["Career"].unique()
```

Regex para eliminar las substrings redundantes

```python
#Licenciatura
df['Career'] = df['Career'].str.replace(r'^Licenciatura\s+en\s+', '', regex=True)
#Especialidad
df['Career'] = df['Career'].str.replace(r'^Especialidad\s+en\s+', '', regex=True)
#Especializaci√≥n
df['Career'] = df['Career'].str.replace(r'^Especializaci√≥n\s+en\s+', '', regex=True)
#Maestr√≠a
df['Career'] = df['Career'].str.replace(r'^Maestr√≠a\s+en\s+', '', regex=True)
#Doctorado
df['Career'] = df['Career'].str.replace(r'^Doctorado\s+en\s+', '', regex=True)
#Ingenier√≠a
df['Career'] = df['Career'].str.replace(r'^Ingenier√≠a\s+en\s+', '', regex=True)
#PUE
df['Career'] = df['Career'].str.replace(r'^Programa\s√∫nico\sde\sespecialidades\s+en\s+', '', regex=True)

df['Career'] = df['Career'].str.replace(r'la\s+ense√±anza', 'Ense√±anza', regex=True)
df['Career'] = df['Career'].str.replace(r'docencia\s+para\s+la\s+educaci√≥n\s+media\s+superior:', 'DEMS', regex=True)

df['Career'].unique()
```
